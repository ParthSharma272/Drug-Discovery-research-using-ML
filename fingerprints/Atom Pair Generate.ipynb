```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
from rdkit import DataStructs
import numpy as np

# Function to generate Atom Pair fingerprint
def generate_atom_pair_fp(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fp = rdMolDescriptors.GetHashedAtomPairFingerprintAsBitVect(mol, nBits=2048)  # 2048 is a typical size
        arr = np.zeros((1,), dtype=int)
        DataStructs.ConvertToNumpyArray(fp, arr)
        return arr.tolist()
    else:
        return None

df['AtomPair'] = df['Smiles'].apply(generate_atom_pair_fp)

# Remove entries with None fingerprints
df = df.dropna(subset=['AtomPair'])

# Convert Atom Pair fingerprints to DataFrame
atom_pair_df = pd.DataFrame(df['AtomPair'].tolist(), columns=[f'AtomPair_{i}' for i in range(len(df['AtomPair'].iloc[0]))])

# Concatenate original DataFrame with Atom Pair fingerprints
combined_df = pd.concat([df.reset_index(drop=True), atom_pair_df.reset_index(drop=True)], axis=1)

# Drop the 'AtomPair' column
combined_df.drop(columns=['AtomPair'], inplace=True)
```
