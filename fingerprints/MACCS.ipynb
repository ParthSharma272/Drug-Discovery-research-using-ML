```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import MACCSkeys
import numpy as np

def generate_maccs_keys(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fp = MACCSkeys.GenMACCSKeys(mol)
        arr = np.zeros((167,), dtype=int)  # MACCS keys have 167 bits
        for i in range(len(fp)):
            arr[i] = fp[i]
        return arr.tolist()
    else:
        return None



# Generate MACCS keys
df['MACCS'] = df['Smiles'].apply(generate_maccs_keys)

# Remove entries with None fingerprints
df = df.dropna(subset=['MACCS'])

# Convert MACCS keys to DataFrame
maccs_df = pd.DataFrame(df['MACCS'].tolist(), columns=[f'MACCS_{i}' for i in range(167)])  # MACCS keys have 167 bits

# Concatenate original DataFrame with MACCS keys
combined_df = pd.concat([df.reset_index(drop=True), maccs_df.reset_index(drop=True)], axis=1)

# Drop the 'MACCS' column
combined_df.drop(columns=['MACCS'], inplace=True)
```
