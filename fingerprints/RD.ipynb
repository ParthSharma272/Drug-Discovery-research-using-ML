```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np

# Function to generate Morgan fingerprint
def generate_morgan_fingerprint(smiles, radius=2, nBits=2048):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits)
        arr = np.zeros((1,), dtype=int)
        DataStructs.ConvertToNumpyArray(fp, arr)
        return arr.tolist()  # Convert numpy array to list
    else:
        return None

# Generate Morgan fingerprints
df['Morgan'] = df['Smiles'].apply(generate_morgan_fingerprint)

# Remove entries with None fingerprints
df = df.dropna(subset=['Morgan'])

# Convert Morgan fingerprints to DataFrame
morgan_df = pd.DataFrame(df['Morgan'].tolist(), columns=[f'Morgan_{i}' for i in range(2048)])

# Concatenate original DataFrame with Morgan fingerprints
combined_df = pd.concat([df.reset_index(drop=True), morgan_df], axis=1)

# Drop the 'Morgan' column
combined_df.drop(columns=['Morgan'], inplace=True)

# Save the combined DataFrame to an Excel file
combined_df.to_excel('combined_dataset_with_morgan.xlsx', index=False)
```
